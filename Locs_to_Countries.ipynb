```python
def loc_to_country(lat, lng):
    query_url = f"https://maps.googleapis.com/maps/api/geocode/json?latlng={lat},{lng}&key={API_KEY}"
    response = req.get(query_url).json()
    pprint(response)
    address_comps = response['results'][0]['address_components']
    for comp in address_comps:
        if (comp['types'][0] == 'country'):
            country = comp['long_name']
    
    return country
```

```python
incidents = pd.read_csv("CSVs/MIDLOCI_2.0.csv", encoding="ISO-8859-1")
incidents.head()
```

```python
incidents = incidents.loc[incidents['year']==2010,:]
incidents.reset_index(drop=True, inplace=True)
incidents.head(10)
```

```python
incidents = incidents.iloc[0:10,:]
```

```python
incident_countries = []

for x in range(len(incidents)):
    row = incidents.iloc[x,:]
    lat = row[6]
    lng = row[5]
    country = loc_to_country(lat, lng)
    incident_countries.append(country)
    
print(incident_countries)

## Can I pull out the long_name for wherever 'address_components':'types', includes['country', 'political']
```
